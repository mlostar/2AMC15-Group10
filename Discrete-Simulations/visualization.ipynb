```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
data = pd.read_csv('/Users/meruozhu/Downloads/monte_carlo_on_policy.csv')
```

```python
data
```

```python
#pd.unique(data['random_move_prob'])
plot_names = pd.unique(data['random_move_prob']).tolist()
```

```python
para_fixed = 'random_move_prob'
```

```python
df=data.groupby([para_fixed]).get_group(0.5)[["mean cleaned percentage","epsilon","gamma"]]
```

```python
Cols = pd.unique(df.iloc[:,1]).tolist()
Index = pd.unique(df.iloc[:,2]).tolist()
n_param_0 = len(Index)
n_param_1 = len(Cols)
score = np.array(df.iloc[:,0])
print(type(score))
score = np.reshape(score,[n_param_0,n_param_1])
```

```python
df_draw = pd.DataFrame(score, index=Index, columns=Cols)
ax=sns.heatmap(df_draw, annot=True)
ax.set(xlabel = df.columns[1], ylabel = df.columns[2])
plot_name = plot_names[0]
title = para_fixed + '=' +str(plot_names[1])
ax.set_title(title)
```

```python
f_name = 'SARSA'
data = pd.read_csv('/Users/meruozhu/Downloads/'+f_name+'.csv')
data
```

```python
plot_names_0 = pd.unique(data['random_move_prob']).tolist()
plot_names_1 = pd.unique(data['alpha']).tolist()
```

```python
para_fixed_0 = 'random_move_prob'
para_fixed_1 = 'alpha'
i=0
j=1
df=data.groupby([para_fixed_0]).get_group(plot_names_0[i])[["mean cleaned percentage","epsilon","gamma","alpha"]]
df=df.groupby([para_fixed_1]).get_group(plot_names_1[j])[["mean cleaned percentage","epsilon","gamma"]]
df
```

```python
para_fixed_0 = 'random_move_prob'
para_fixed_1 = 'alpha'
for i in range(len(plot_names_0)):
    for j in range(len(plot_names_1)):
        df=data.groupby([para_fixed_0]).get_group(plot_names_0[i])[["mean cleaned percentage","epsilon","gamma","alpha"]]
        df=df.groupby([para_fixed_1]).get_group(plot_names_1[j])[["mean cleaned percentage","epsilon","gamma"]]
        Cols = pd.unique(df.iloc[:,1]).tolist()
        Index = pd.unique(df.iloc[:,2]).tolist()
        n_param_0 = len(Index)
        n_param_1 = len(Cols)
        score = np.array(df.iloc[:,0])
        print(type(score))
        score = np.reshape(score,[n_param_0,n_param_1])
        df_draw = pd.DataFrame(score, index=Index, columns=Cols)
        ax=sns.heatmap(df_draw, annot=True)
        ax.set(xlabel = df.columns[1], ylabel = df.columns[2])
        title = para_fixed_0 + '=' +str(plot_names_0[i]) + "," + para_fixed_1 + '=' +str(plot_names_1[j])
        ax.set_title(title)
        fig = ax.get_figure()
        fig.savefig('/Users/meruozhu/Downloads/2AMC15/2AMC15-Group10/Discrete-Simulations/heatmaps/' + f_name + '_' + para_fixed_0 + '_' +str(plot_names_0[i]) + "_" + para_fixed_1 + '_' +str(plot_names_1[j])+'.jpg')
        fig.clf()
```
